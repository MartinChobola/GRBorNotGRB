```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import glob

from astropy.time import Time
```

```python
uno = glob.glob("Data/GRBalpha/Data_firmware_2/*-*-*_*-*-*_exp*s_sbin*/ch0/part*/rate_multi_ch.txt")
dos = glob.glob("Data/GRBalpha/Data_firmware_3/*-*-*_*-*-*_exp*s_sbin*/ch0/rate_multi_ch.txt")
data = np.sort(uno+dos)
storage = []
for fname in data:
    with open(fname) as file:
        lines = file.readlines()
        for line in lines[9:]:
            split = line[:-1].split()            
            times = [split[2]+' '+split[3]]
            t = Time(times, format='iso', scale='utc')
            storage.append([float(t.jd - 2459321), float(split[7]),float(split[8]), float(split[9]), float(split[-1])])
```

```python
df = pd.DataFrame(storage, columns=["time", "lon","lat","alt","cps"])
df
```

```python
df.to_parquet("GRBalpha_all.parquet")
```
